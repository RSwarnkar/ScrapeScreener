# Summerize Key Ratios: 

```python
import requests as rq
import pandas as pd 
import time
import datetime
```

## Begin Code: 

```python
import csv
company_list = []

with open('screener-stocks-list.csv', newline='') as f:
    for row in csv.reader(f):
         company_list.append(row[0])

company_list.pop(0)
#print(company_list)
```

## Sales Growth:

```python
sale_growth_df = pd.DataFrame(columns=['Stock','10 Years', '5 Years', '3 Years', 'TTM'])

for i in range(len(company_list)):
    url = "https://www.screener.in/company/"+company_list[i]+"/consolidated/"
    print("Loading Page for ",company_list[i])
    tables = pd.read_html(url)
    time.sleep(1) # Seconds
    temp_df = tables[2]
    temp_list = []
    temp_list.append(company_list[i]) 
    temp_list = temp_list + temp_df['Compounded Sales Growth.1'].tolist()
    #print(temp_list)
    sale_growth_df.loc[i] =  temp_list

filename = "output/sales-growth-"+datetime.datetime.now().strftime("%d-%m-%Y-%H-%M-%S")+".csv"
sale_growth_df.to_csv(filename,index=False)

sale_growth_df.tail(5)
```
